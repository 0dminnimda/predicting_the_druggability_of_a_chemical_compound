```python
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit import DataStructs
from rdkit.Chem import AllChem, Draw, Descriptors
from rdkit.Chem.Draw import IPythonConsole
from sklearn.preprocessing import FunctionTransformer
```

```python
def RDKfingerPrint(mol_smi, **kwargs):
    mol = Chem.MolFromSmiles(mol_smi)
    desc_vec = np.zeros((1,), dtype=int)
    DataStructs.ConvertToNumpyArray(AllChem.RDKFingerprint(mol, **kwargs), desc_vec)
    return desc_vec


def mol_fingerprint_calc(mols):
    return pd.DataFrame(RDKfingerPrint(m, maxPath=5) for m in mols)


descriptors_transformer = FunctionTransformer(mol_fingerprint_calc)
X = descriptors_transformer.transform(data['SMILES'])
```

```python
X
```

```python
data = pd.concat([data, X], axis=1)
data
```

```python
data.to_csv(name + '_emb.csv', index=False)
```
